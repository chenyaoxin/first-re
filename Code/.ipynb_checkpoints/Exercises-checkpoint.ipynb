# Exercise1

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([203,58,210,202,198,158,165,201,157,131,166,160,186,125,218,146])
y=np.array([495,173,479,504,510,416,393,442,317,311,400,337,423,334,533,344])
yerr=np.array([21,15,27,14,30,16,14,25,52,16,34,31,42,26,16,22])
A=np.vander(x,2)
C=np.diag(yerr*yerr)
ATA=np.dot(np.dot(A.T,np.linalg.inv(C)),A)
cov=np.linalg.inv(ATA)
ATY=np.dot(np.dot(A.T,np.linalg.inv(C)),y)
W=np.dot(np.linalg.inv(ATA),ATY)
print("m={0:.2f}±{1:.2f}".format(W[0],np.sqrt(cov[0,0])))
print("b={0:.0f}±{1:.0f}".format(W[1],np.sqrt(cov[1,1])))

plt.errorbar(x,y,yerr=yerr,fmt='.k',ecolor='k',elinewidth=0,capsize=4)
plt.text(165,200,"y=({0:.2f}±{1:.2f})x+({2:.0f}±{3:.0f})".format(W[0],np.sqrt(cov[0,0]),W[1],np.sqrt(cov[1,1])))
plt.xlim(0, 300)
plt.ylim(0,700)

x0 = np.linspace(0, 300, 10)
plt.plot(x0,np.dot(np.vander(x0,2),W),"k")

plt.xlabel("x")
plt.ylabel("y")
```

# Exercise3

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([203,58,210,202,198,158,165,201,157,131,166,160,186,125,218,146])
y=np.array([495,173,479,504,510,416,393,442,317,311,400,337,423,334,533,344])
yerr=np.array([21,15,27,14,30,16,14,25,52,16,34,31,42,26,16,22])
A=np.vander(x,3)
C=np.diag(yerr*yerr)
ATA=np.dot(np.dot(A.T,np.linalg.inv(C)),A)
cov=np.linalg.inv(ATA)
ATY=np.dot(np.dot(A.T,np.linalg.inv(C)),y)
W=np.dot(np.linalg.inv(ATA),ATY)
print("q={0:.4f}±{1:.4f}".format(W[0],np.sqrt(cov[0,0])))
print("m={0:.2f}±{1:.2f}".format(W[1],np.sqrt(cov[1,1])))
print("b={0:.0f}±{1:.0f}".format(W[2],np.sqrt(cov[2,2])))

plt.errorbar(x,y,yerr=yerr,fmt='.k',ecolor='k',elinewidth=0,capsize=4)
plt.text(75,100,"y=(({0:.4f}±{1:.4f})$x^2$+({2:.2f}±{3:.2f})x+({4:.0f}±{5:.0f})".format(W[0],np.sqrt(cov[0,0]),W[1],np.sqrt(cov[1,1]),W[2],np.sqrt(cov[2,2])))
plt.xlim(0, 300)
plt.ylim(0,700)

x0 = np.linspace(0, 300, 10)
plt.plot(x0,np.dot(np.vander(x0,3),W),"k")

plt.xlabel("x")
plt.ylabel("y")
```
